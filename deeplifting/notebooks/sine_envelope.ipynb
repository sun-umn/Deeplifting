```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def sine_envelope1(x1, x2):
    numerator = np.square(np.sin(np.sqrt(x2**2 + x1**2))) - 0.5
    denominator = np.square(0.001 * (x2**2 + x1**2) + 1)
    result = (numerator / denominator) + 0.5
    return result
```

```python
def sine_envelope2(x1, x2):
    numerator = np.square(np.sin(np.sqrt(x2**2 + x1**2) - 0.5))
    denominator = np.square(0.001 * (x2**2 + x1**2) + 1)
    result = (numerator / denominator) + 0.5
    return result
```

```python
x1_min, x1_max = -20, 20
x1_axis = np.arange(x1_min, x1_max, 0.1)

x2_min, x2_max = -20, 20
x2_axis = np.arange(x2_min, x2_max, 0.1)
x1, x2 = np.meshgrid(x1_axis, x2_axis)
f = sine_envelope1(x1, x2)

figure = plt.figure()
axis = figure.add_subplot(111, projection="3d")
axis.plot_surface(x1, x2, f, cmap="jet", shade="false")

# Set the title
axis.set_title('Sine Envelope with -0.5 outside Sine')
```

```python
x1_min, x1_max = -20, 20
x1_axis = np.arange(x1_min, x1_max, 0.1)

x2_min, x2_max = -20, 20
x2_axis = np.arange(x2_min, x2_max, 0.1)
x1, x2 = np.meshgrid(x1_axis, x2_axis)
f = sine_envelope2(x1, x2)

figure = plt.figure()
axis = figure.add_subplot(111, projection="3d")
axis.plot_surface(x1, x2, f, cmap="jet", shade="false")

# Set the title
axis.set_title('Sine Envelope with -0.5 Inside Sine')
```
