# Griewank Function
The Griewank function is complex optimization problem with many regularly spread local minima and a single global minimum.
It's mathematical formulation is $$f(\textbf{x})=\sum_{i=1}^d\frac{x_i^2}{4000}-\prod_{i=1}^d\cos{\frac{x_i}{\sqrt{i}}}+1$$
where d is the number of dimensions. 

## Input Domain

The function is typically evaluated on the region $x_i\in[-600,600], \forall i=1,\dots,d.$

## Global Minimum

It's global minimum is $f(\textbf{x}^*)=0,$ located at $\textbf{x}^*=(0,\dots,0).$

# Function Visualization

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Objective function in 3d
def griewank(x1,x2):
    return (1/4000)*(x1**2+x2**2)-(np.cos(x1)*np.cos(x2/(2**0.5)))+1

print(griewank(0,0))
```

```python
# Plot Drop-Wave in 3D
r_min, r_max = -600, 600
x_axis = y_axis = np.arange(r_min, r_max, 5)
x1, x2 = np.meshgrid(x_axis, y_axis)
f = griewank(x1, x2)

figure = plt.figure()
axis = figure.add_subplot(111, projection="3d")
axis.plot_surface(x1, x2, f, cmap="jet", shade="false")

# Set the title
axis.set_title("Griewank Surface Plot (1200x1200)")
```

```python
# Let's zoom in to see more local minima
r_min, r_max = -50, 50
x_axis = y_axis = np.arange(r_min, r_max, 1)
x1, x2 = np.meshgrid(x_axis, y_axis)
f = griewank(x1, x2)

figure = plt.figure()
axis = figure.add_subplot(111, projection="3d")
axis.plot_surface(x1, x2, f, cmap="jet", shade="false")

# Set the title
axis.set_title("Griewank Surface Plot (100x100)")
```
